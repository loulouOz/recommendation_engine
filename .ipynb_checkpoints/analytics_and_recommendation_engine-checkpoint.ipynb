## Using RFM Segmentation (Recency, Frequency and Monetary Value Analysis) to recommend other purchases:

### First step:
    
**RFM segmentation is a great method to identify groups of customers for special treatment. Excellent method to improve your customer marketing.**

While there are countless ways to perform segmentation, RFM analysis is popular for three reasons:

It utilizes objective, numerical scales that yield a concise and informative high-level depiction of customers.
It is simple – marketers can use it effectively without the need for data scientists or sophisticated software.
It is intuitive – the output of this segmentation method is easy to understand and interpret.

- **Recency:** How much time has elapsed since a customer’s last activity or transaction with the brand? (Activity could be the last visit to a website or use of a mobile app, etc...)

- **Frequency:** How often has a customer transacted or interacted with the brand during a particular period of time? Clearly, customers with frequent activities are more engaged, and probably more loyal, than customers who rarely do so. And one-time-only customers are in a class of their own.

- **Monetary:** Also referred to as “monetary value,” this factor reflects how much a customer has spent with the brand during a particular period of time. Big spenders should usually be treated differently than customers who spend little. Looking at monetary divided by frequency indicates the average purchase amount – an important secondary factor to consider when segmenting customers.


### Second step:

Using Singular Value Decomposition (SVD) to recommend purchases using our best customers identified in the above step for better targeting.

SVD is a classical Collaborative Filtering method from linear algebra popular for developing recommender systems.

#### What Is Collaborative Filtering?

Collaborative filtering is a technique that can filter out items that a user might like on the basis of reactions by similar users.

It works by searching a large group of people and finding a smaller set of users with tastes similar to a particular user. It looks at the items they like and combines them to create a ranked list of suggestions.


### Third step:

Using Dimensionality reduction using truncated Singular Value Decomposition (truncatedSVD) with a correlation matrix to recommend other products that customers also bought with a specific product. AKA The "people who bought this also bought..." that you see in your cart before finalising your purchase.


### Load and prepare dataset:

```python
df = pd.read_csv('e-commerce_data.csv',encoding = "ISO-8859-1")
print('rows:',df.shape[0],'columns:',df.shape[1])

df['TotalPrice'] = df['Quantity']*df['UnitPrice']
df['CustomerID'] = df['CustomerID'].fillna(0).astype(int).astype(str).replace('0',np.nan)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['day'] = df['InvoiceDate'].dt.day
df['day_name'] = df['InvoiceDate'].dt.day_name() 
df['day_of_year'] = df['InvoiceDate'].dt.dayofyear 
df['week'] = df['InvoiceDate'].dt.weekofyear 
df['hour'] = df['InvoiceDate'].dt.hour  
df['month'] = df['InvoiceDate'].dt.month
df['year'] = df['InvoiceDate'].dt.year
df['weekday'] = df['InvoiceDate'].dt.dayofweek
df['weekend'] = [0 if x <= 5 else 1 for x in df['weekday']]

df['month_year'] = df['year'].map(str)+ '-' +df['month'].map(str)
df['month_year'] = pd.to_datetime(df['month_year'], format='%Y-%m').dt.strftime('%Y-%m')

df.head(3)
```

```python
def resume_table(df):
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name','dtypes']]
    summary['Missing'] = df.isnull().sum().values 
    summary['Missing_%'] = np.round(df.isnull().sum().values / len(df) *100,2)
    summary['Uniques'] = df.nunique().values
    summary['First 5 Unique Values'] = [x for x in pd.Series({c: df[c].unique()[:5] for c in df})]
    
    ## Entropy is the measures of impurity, disorder or uncertainty.
    ## Entropy controls how a Decision Tree decides to split the data. It actually effects how a Decision Tree draws its boundaries.
    ## An attribute with highest Information gain will tested/split first.
    for name in summary['Name'].value_counts().index:
        summary.loc[summary['Name'] == name, 'Entropy'] = round(stats.entropy(df[name].value_counts(normalize=True), base=2),2) 

    return summary

df_resume = resume_table(df)
df_resume
```

```python
# df[df['InvoiceNo'].str.contains('C')]['InvoiceNo'].unique()
df = df[~df['InvoiceNo'].str.contains('C')]
df = df[df['Description'] != 'POSTAGE']
```

### Columns 'Description' and 'CustomerID' have missing values:

- 'Description' feature not only have empty descriptions but also incorrect ones (same stockcode with different descriptions), so instead of just dropping them we will map them back based on their relation 'StockCode' to 'Description'.

- 'CustomerID' feature has 3710 invoice with no CustomerID which represents 135 thousand lines, I will filter them out for this exercise.

#### Treatment of the 'Description' column:

```python
df['Description'] = df['Description'].str.strip()

df['StockCode'] = df['StockCode'].str.upper()

## description that comes back the most per 'StockCode'. if a list comes back, choose first element.
stockCode = df.groupby(['StockCode'])['Description'].agg(lambda x: x.mode(dropna=False).iloc[0]).to_frame().rename(columns={'Description':'Description2'}) #.agg(pd.Series.mode).rename('Description2')

print('Empty descriptions before fix:',len(df[df['Description'].isnull()]))
df = df.merge(stockCode,how='left',left_on='StockCode',right_on='StockCode').drop('Description',axis=1).rename(columns={'Description2':'Description'})
df['Description'] = df['Description'].astype(str).apply(lambda x: np.nan if len(x)==0 else x)
print('Empty descriptions after fix:',len(df[df['Description'].isnull()]))
# df.head(3)
```

```python
print(len(df))
df = df[~df['CustomerID'].isnull()]
print(len(df))
```

### 1. RFM Segmentation - Recency, Frequency and Monetary Value Analysis:

```python
df1 = df.groupby(['CustomerID']).agg({'InvoiceDate':[min,max],'InvoiceNo':pd.Series.nunique,'TotalPrice':np.sum}).reset_index()
df1.columns = [' '.join(col).strip() for col in df1.columns.values]
df1 = df1.rename(columns={'InvoiceDate min':'FirstInvoice','InvoiceDate max':'LastInvoice','InvoiceNo nunique':'NumInvoice','TotalPrice sum':'TotalSales'})
## DaysBetweenFirstLastInv
df1['DaysBetweenFirstLastInv'] = (df1['LastInvoice']-df1['FirstInvoice']).astype('timedelta64[D]') # 1 +
## days_as_customer
df1['lastDate'] = df['InvoiceDate'].max()
df1['days_as_customer'] = (df1['lastDate']-df1['FirstInvoice']).astype('timedelta64[D]') #1 + 
## days_since_purchase
df1['days_since_purchase'] = (df1['lastDate']-df1['LastInvoice']).astype('timedelta64[D]') #1 + 
## Average days between purchases:
df1['AvgDaysBetweenPurchases'] = df1['DaysBetweenFirstLastInv'] / (df1['NumInvoice']-1)
## Days Above Average: 'days_since_purchase' -  'AvgDaysBetweenPurchases' or 'NextPurchaseInDays': 'AvgDaysBetweenPurchases' - 'days_since_purchase'
df1['NextPurchaseInDays'] = df1['AvgDaysBetweenPurchases'] - df1['days_since_purchase']

df1 = df1.sort_values('NextPurchaseInDays',ascending=False)

display_side_by_side(df1.head(),titles=['When Will Your Customers Purchase Next?'])
```

```python
df1.drop(['FirstInvoice','LastInvoice','DaysBetweenFirstLastInv','days_as_customer','lastDate','AvgDaysBetweenPurchases','NextPurchaseInDays'],axis=1, inplace=True)

rfm = df1[['CustomerID','days_since_purchase','NumInvoice','TotalSales']].copy().sort_values('NumInvoice', ascending=False)

rfm['days_since_purchase_Q'] = pd.qcut(rfm['days_since_purchase'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['NumInvoice_Q'] = pd.qcut(rfm['NumInvoice'].rank(method='first'), 4, labels=[4,3,2,1])
rfm['TotalPrice_Q'] = pd.qcut(rfm['TotalSales'].rank(method='first'), 4, labels=[4,3,2,1])

rfm['segment'] = rfm['days_since_purchase_Q'].astype(str) + "-" + rfm['NumInvoice_Q'].astype(str) + "-" + rfm['TotalPrice_Q'].astype(str)
rfm2 = rfm.copy()
rfm.drop(['days_since_purchase_Q','NumInvoice_Q','TotalPrice_Q'], axis=1,inplace=True)

rfm.head(3)
```

```python
Best_Customers = rfm[rfm['segment'] == '1-1-1'].sort_values('TotalSales',ascending=False)

listy = ['1-4-1','1-4-2','1-3-1','1-3-2']
High_spending_New_Customers = rfm[rfm['segment'].isin(listy)].sort_values('TotalSales',ascending=False)

listy = ['1-1-3','1-1-4']
Lowest_Spending_Active_Loyal_Customers = rfm[rfm['segment'].isin(listy)].sort_values('NumInvoice',ascending=False)

listy = ['4-1-1','4-1-2','4-2-1','4-2-2']
Churned_Best_Customers = rfm[rfm['segment'].isin(listy)].sort_values('TotalSales',ascending=False)

display_side_by_side(Best_Customers.head(), High_spending_New_Customers.head(),
                     Lowest_Spending_Active_Loyal_Customers.head(),Churned_Best_Customers.head(),
                     titles=['Best_Customers - count= {}'.format(len(Best_Customers)),'High_spending_New_Customers - count= {}'.format(len(High_spending_New_Customers)),
                             'Lowest_Spending_Active_Loyal_Customers - count= {}'.format(len(Lowest_Spending_Active_Loyal_Customers)),'Churned_Best_Customers - count= {}'.format(len(Churned_Best_Customers))])
```

```python
listo = ['days_since_purchase','NumInvoice','TotalSales']

fig,ax = plt.subplots(1,3,figsize=(20,4))
for i in range(len(listo)):
    sns.distplot(Best_Customers[listo[i]], hist=True, ax=ax[i]);
```

#### Let's merge back to keep our best customers (segment 1-1-1):

```python
bestCustomer = df.merge(Best_Customers,how='left',left_on='CustomerID',right_on='CustomerID')
bestCustomer = bestCustomer[~bestCustomer['segment'].isnull()]
print(len(bestCustomer))
bestCustomer.head(3)
```

### 2. Singular Value Decomposition (SVD) to recommend purchases using our best customers:

#### Here, we do not have a rating column so let's create one by looking at how many times a product has been purchased (# of invoices) divided by the total number of invoices for each customer and create bins for rating between 1 to 5.
(this is why I keep the best customers only to be more meaningful: more invoices, customers for a while, etc...)

```python
freq = bestCustomer.groupby(['CustomerID','Description']).agg({'InvoiceNo':pd.Series.nunique,'NumInvoice':np.unique}).reset_index()
freq['rating'] = freq['InvoiceNo'] / freq['NumInvoice']

freq['rating'] = pd.cut(freq['rating'], bins=[0,0.2,0.4,0.6,0.8,1], labels=[1,2,3,4,5])
freq['rating'] = freq['rating'].astype(int)

print(len(freq),min(freq['rating']),max(freq['rating']))
freq.head(3)
```

#### Let's create also a unique column of descriptions of purchases to use below with the SVD:

```python
desc = pd.DataFrame()
desc['Description'] = freq['Description'].unique()
print(len(desc))
desc.head(3)
```

#### Import packages used for the SVD:

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection.validation import cross_validate
from surprise.model_selection import GridSearchCV
```

**The Dataset module in Surprise provides different methods for loading data from files, Pandas DataFrames, or built-in datasets:**

Dataset.load_builtin()
Dataset.load_from_file()
Dataset.load_from_df()

The **Reader class** in Surprise is to parse a file containing users, items, and users’ ratings on items. The default format is that each rating is stored in a separate line in the following order separated by space: user item rating
This order and the separator are configurable using the following parameters:

- **line_format** is a string like “item user rating” to indicate the order of the data with field names separated by a space
- **sep** is used to specify separator between fields, such as space, ‘,’, etc.
- **rating_scale** is to specify the rating scale. The default value is (1, 5)
- **skip_lines** is to indicate the number of lines to skip at the beginning of the file and the default is 0

let's load the dataframe and create a SVD model instance:

```python
reader = Reader(rating_scale=[1,5])

data = Dataset.load_from_df(freq[['CustomerID','Description','rating']], reader)
svd = SVD()
```

### Hypertuning of SVD:
(trying multiple combinations until you fin the best/optimised parameters)

```python
%%time

param_grid = {
              "n_factors":[22],
              "n_epochs": [38],
              "lr_all": [0.004],
              "reg_all": [0.1]
                }
gs = GridSearchCV(SVD, param_grid, measures=["rmse", "mae"], cv=3)

gs.fit(data)

print(gs.best_score["rmse"])
print(gs.best_params["rmse"])

params = gs.best_params['rmse']
svdtuned = SVD(n_factors=params['n_factors'], 
               n_epochs=params['n_epochs'],
               lr_all=params['lr_all'], 
               reg_all=params['reg_all'])
```

```python
# # Run 5-fold cross-validation and print results
# cross_validate(svdtuned, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

#### Once the model has been evaluated to our satisfaction, then we can re-train the model using the entire training dataset:

```python
trainset = data.build_full_trainset()
svdtuned.fit(trainset)
```

### Recommending Products:

After the recommendation model has been trained appropriately, it can be used for prediction.
For example, given a customer, we can use the trained model to predict products in which the customer may be interested.

```python
## For one customer:
customer = '16360' #'15993' #'14911'

descrip = desc.copy()

descrip['Estimate_Score'] = descrip['Description'].apply(lambda x: svdtuned.predict(customer, x).est)
descrip = descrip.sort_values(by=['Estimate_Score'], ascending=False)
print(len(descrip))
display_side_by_side(descrip.head(10),titles=['Product recommendations for customer #{}'.format(customer)])
```

### 3. Market-Basket Analysis also known as 'people who bought this also bought...':

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

## shape the 'rating' dataset into a matrix
basket = df.groupby(['InvoiceNo', 'Description']).agg({'Quantity':np.sum}).unstack().reset_index().fillna(0).set_index('InvoiceNo')
## replace values bigger than 1 by one:
basket[basket>1]=1
## fix columns name:
basket.index.name = None
basket.columns = [col[1] for col in basket.columns]
## Transpose
basket = basket.T
# print(basket.head())

## Dimensionality reduction using truncated SVD (aka LSA). 
## This transformer performs linear dimensionality reduction by means of truncated singular value decomposition:
TruncSVD = TruncatedSVD(n_components=100)
decomposed = TruncSVD.fit_transform(basket)
print('decomposed matrix shape:',decomposed.shape)

## Correlation matrix:
correlation = np.corrcoef(decomposed)
print('correlation matrix shape:',correlation.shape)
```

```python
# import random

# ## Choose a random product:
# foo = freq['Description'].unique()
# product = random.choice(foo)
# print(product)

product = 'GARDENERS KNEELING PAD CUP OF TEA'
```

```python
list_descriptions = list(basket.index)
product_index = list_descriptions.index(product)
print('Index of product:',product_index)

correlation_product_index = correlation[product_index]
print('Number of correlations of product:',correlation_product_index.shape)
```

```python
threshold = 0.7
Recommend = pd.DataFrame(list([(x,y) for x,y in zip(basket.index[correlation_product_index > threshold],
                                                    correlation_product_index[correlation_product_index > threshold])]),columns=['Description','Correlation'])
Recommend = Recommend[Recommend['Description'] != product].sort_values('Correlation',ascending=False)
print('Number of related products above threshold:',len(Recommend))
display_side_by_side(Recommend.head(),titles=['People who bought {}, also bought:'.format(product)])
```
